```python
import PyPDF2
import pandas as pd
import glob, os

# Iniciamos las listas
acc = []
toam = []
bida = []
names = []
menu = []  # Mantenemos la lista 'menu' fuera del bucle

ext = "*.pdf"
PATH = "/content/drive/MyDrive/Prueba PDF"

# Obtener todos los archivos PDF en el directorio
all_files = []
for dirpath, dirnames, filenames in os.walk(PATH):
    all_files += glob.glob(os.path.join(dirpath, ext))

for idx, filename in enumerate(all_files):
    try:
        pdf_reader = PyPDF2.PdfReader(filename)
        # Trabajar con la primera hoja del pdf
        page = pdf_reader.pages[0]
        # Extraer texto
        text = page.extract_text()
        # Dividir el texto en lineas
        lines = text.split('\n')

        # Agregar el nombre del archivo a la lista
        names.append(filename)

        # Variables temporales
        accountNumber = ''
        totalAmount = ''
        billDate = ''
        previous = ''  # Usamos 'previous' en lugar de 'menu'

        for line in lines:
            if 'Account Number' in line:
                accountNumber = line.split('Account Number')[-1].strip()
            elif 'Total Amount Due' in line:
                totalAmount = line.split('Total Amount Due')[-1].strip()
            elif 'Bill Date' in line:
                billDate = line.split('Bill Date')[-1].strip()
            elif 'Read Dates' in line:
                previous = line.split('Read Dates')[-1].strip()

        # Agregar a las listas
        acc.append(accountNumber)
        toam.append(totalAmount)
        bida.append(billDate)
        menu.append(previous)  # Asegurarse de agregar 'previous' a 'menu'

    except Exception as e:
        print(f"Error procesando el archivo {filename}: {e}")

# Crear el DataFrame
data = {'Files': names, 'Account Number': acc, 'Total Amount Due': toam, 'Bill Date': bida, 'Previous': menu}
df = pd.DataFrame(data)
df
```
